# Vectors - Projections and Divisions


## Projecting a Vector onto Another Vector

Projecting vector $\mathbf{a}$ onto axis $\mathbf{b}$ involves finding the component of vector $\mathbf{a}$ along the direction of $\mathbf{b}$.

Projection of $\mathbf{a}$ onto $\mathbf{b}$:

$$
\text{Proj}_{\mathbf{b}} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{b}}{|\mathbf{b}|^2} \mathbf{b}
$$


## Projecting a Vector onto Axes

Projecting vector $\mathbf{a}$ onto an axis involves finding the component of vector $\mathbf{a}$ along that axis.

Projection of $\mathbf{a}$ onto the $OX$ axis:

$$
\text{Proj}_{OX} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{i}}{|\mathbf{i}|^2} \mathbf{i} = \frac{\mathbf{a} \cdot \mathbf{i}}{1} \mathbf{i} = (\mathbf{a} \cdot \mathbf{i}) \mathbf{i}
$$

Projection of $\mathbf{a}$ onto the $OY$ axis:

$$
\text{Proj}_{OY} \mathbf{a} = \frac{\mathbf{a} \cdot \mathbf{j}}{|\mathbf{j}|^2} \mathbf{j} = \frac{\mathbf{a} \cdot \mathbf{j}}{1} \mathbf{j} = (\mathbf{a} \cdot \mathbf{j}) \mathbf{j}
$$

Recall that $\mathbf{i} = (1, 0)$ and $\mathbf{j} = (0, 1)$ are unit vectors along the $OX$ and $OY$ axes.


## Dividing a Line Segment in a Given Ratio

A line segment is defined by two points in space.

Line segments and vectors are closely related because a line segment connecting two points can be treated as a vector directed from one point to another.

If point \( P \) divides the line segment connecting points $A(x_1, y_1)$ and $B(x_2, y_2)$ in the ratio $m:n$, then the coordinates of point $P$ can be found using the formula:

$$
P = \left(\frac{n \times x_2 + m \times x_1}{m+n}, \frac{n \times y_2 + m \times y_1}{m+n}\right)
$$

**Example:**

Given a line segment connecting points $A(1, 2)$ and $B(3, 8)$, find point $P$ that divides the segment in the ratio 2:3.

Coordinates of point $P$:
$$
P = \left(\frac{2 \times 3 + 3 \times 1}{2+3}, \frac{2 \times 8 + 3 \times 2}{2+3}\right) = \left(\frac{9}{5}, \frac{22}{5}\right)
$$

```python
import matplotlib.pyplot as plt
import math

# Coordinates of points A and B
A = (1, 2)
B = (3, 8)

# Ratio m:n
m, n = 2, 3

# Calculate coordinates of point P that divides the segment in the ratio 2:3
P_x = (n * A[0] + m * B[0]) / (m + n)
P_y = (n * A[1] + m * B[1]) / (m + n)
P = (P_x, P_y)

# Plotting the graph
plt.figure(figsize=(6, 6))

# Plotting line segment AB
plt.plot([A[0], B[0]], [A[1], B[1]], 'k--', label='Line Segment AB')

# Marking points A, B, and P
plt.scatter(*A, color='blue', label='A(1, 2)')
plt.scatter(*B, color='red', label='B(3, 8)')
plt.scatter(*P, color='green', label=f'P({P[0]:.2f}, {P[1]:.2f})')

# Point labels
plt.text(A[0]-0.2, A[1], 'A(1, 2)', fontsize=12, color='blue')
plt.text(B[0]+0.1, B[1], 'B(3, 8)', fontsize=12, color='red')
plt.text(P[0]+0.1, P[1], f'P({P[0]:.2f}, {P[1]:.2f})', fontsize=12, color='green')

# Adding grid and axis labels
plt.grid(True)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Dividing a Line Segment in the Ratio 2:3')
plt.legend()

# Displaying the graph
plt.show()
```

Project vector  a=(3,4)  onto the  OX  and  OY  axes. Illustrate the result on a graph.

$$
\begin{align*}
OX = \frac{(3, 4) * (1, 0)}{(1, 0)^2} * (1, 0) = \frac{3}{1} * (1, 0) = (3, 0)\\
OY = \frac{(3, 4) *(0, 1)}{(0, 1)^2} * (0, 1) = \frac{4}{4} * (0, 1) = (0, 4)
\end{align*}
$$

```python
import matplotlib.pyplot as plt
import numpy as np

# Define vector a and origin
a = (3, 4)
origin = (0, 0)

# Calculate projections onto OX and OY axes
proj_OX = (a[0], 0)  # Projection onto OX: (3, 0)
proj_OY = (0, a[1])  # Projection onto OY: (0, 4)

# Create the plot
plt.figure(figsize=(6, 6))

# Plot vector a
plt.quiver(*origin, *a, angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a (3, 4)')

# Plot projection onto OX
plt.quiver(*origin, *proj_OX, angles='xy', scale_units='xy', scale=1, color='red',  label='Projection on OX (3, 0)')
plt.plot([origin[0], proj_OX[0]], [origin[1], proj_OX[1]], 'r--')


# Plot projection onto OY
plt.quiver(*origin, *proj_OY, angles='xy', scale_units='xy', scale=1, color='green', label='Projection on OY (0, 4)')
plt.plot([origin[0], proj_OY[0]], [origin[1], proj_OY[1]], 'g--')


# Customize the plot
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.xlim(-1, 4)
plt.ylim(-1, 5)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Projection of Vector (3, 4) onto OX and OY Axes')
plt.legend()

# Show the plot
plt.show()
```

Project vector  a=(2,3)  onto vector  b=(1,1) . Illustrate the result on a graph.

$$
\begin{align*}
= \frac{(2 * 1) + (3 * 1) = 5}{(1 * 1) + (1 * 1) = 2} * (1, 1) = (\frac{5}{2}, \frac{5}{2})
\end{align*}
$$

```python
import matplotlib.pyplot as plt

# Define vectors a and b
a = (2, 3)
b = (1, 1)

# Compute the dot products
dot_ab = a[0] * b[0] + a[1] * b[1]  # a  b
dot_bb = b[0] * b[0] + b[1] * b[1]  # b  b

# Calculate projection of a onto b
proj_b_a = ((dot_ab / dot_bb) * b[0], (dot_ab / dot_bb) * b[1])

# Origin for plotting
origin = (0, 0)

# Plotting the vectors
plt.figure(figsize=(6, 6))

# Plot vector a
plt.quiver(*origin, *a, angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a (2, 3)')

# Plot vector b
plt.quiver(*origin, *b, angles='xy', scale_units='xy', scale=1, color='green', label='Vector b (1, 1)')

# Plot projection of a onto b
plt.quiver(*origin, *proj_b_a, angles='xy', scale_units='xy', scale=1, color='red', label=f'Projection of a onto b ({proj_b_a[0]:.2f}, {proj_b_a[1]:.2f})')

# Adding labels and grid
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Projection of Vector a onto Vector b')
plt.legend()
plt.show()
```

Project vector  b=(1,1)  onto vector  a=(2,3) . Illustrate the result on a graph.

$$
\frac{(1 * 2) + (1 * 3) = 5}{(2 * 2) + (3 * 3) = 13} * (2, 3) =(\frac{10}{13}, \frac{15}{13})
$$

```python
import matplotlib.pyplot as plt

# Define vectors a and b
a = (2, 3)
b = (1, 1)

# Compute the projection of b onto a
dot_ba = b[0] * a[0] + b[1] * a[1]  # b  a
dot_aa = a[0] * a[0] + a[1] * a[1]  # a  a
scalar = dot_ba / dot_aa

proj_a_b = (scalar * a[0], scalar * a[1])

# Origin for plotting
origin = (0, 0)

# Plotting the vectors
plt.figure(figsize=(6, 6))

# Plot vector a
plt.quiver(*origin, *a, angles='xy', scale_units='xy', scale=1, color='blue', label='Vector a (2, 3)')

# Plot vector b
plt.quiver(*origin, *b, angles='xy', scale_units='xy', scale=1, color='green', label='Vector b (1, 1)')

# Plot projection of b onto a
plt.quiver(*origin, *proj_a_b, angles='xy', scale_units='xy', scale=1, color='red', label=f'Projection of b onto a ({proj_a_b[0]:.2f}, {proj_a_b[1]:.2f})')

# Adding labels and grid
plt.axhline(0, color='black', linewidth=0.5, linestyle='--')
plt.axvline(0, color='black', linewidth=0.5, linestyle='--')
plt.xlim(-1, 4)
plt.ylim(-1, 4)
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Projection of Vector b onto Vector a')
plt.legend()
plt.show()
```
