## 20. Limits of Real Functions

1. Compute:
   - $\displaystyle\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3}$

2. Find:
   
   - $\displaystyle \lim_{x \to 0} \frac{\sin(3x)}{2x+1}$.

4. Find the asymptotes of the function:
  
   - $f(x) = \frac{x^2 - 1}{x^2 + 1}$
   - $g(x) = \frac{\sin(x)}{x^2+1}$


   \
   \
   \
   $$
\displaystyle\lim_{x \to \infty} \frac{x^3 + 2x^2}{x^4 - 3x^3} = \frac{x^3(1 + \frac{2}{x})}{x^4 (1 - \frac{3}{x})}
   $$
Instead of simplifying the expressions using the lowest power of x, we use the highest power
\
$$
\frac{x^3(1 - 0)}{x^4(1 - 0)} = \frac{1}{x} = 0
$$
\
\
\
\
I'll be using the range function of SIN for this one

$$
-1 \leq \sin(3x) \leq 1
$$

$$
\frac{-1}{2x + 1} = 0 \leq \frac{\sin(3x)}{2x + 1} \leq \frac{1}{2x + 1} = 0
$$
I used x -> 0 for this one
$$
\begin{align*}
0 \leq \frac{\sin(3x)}{2x + 1} \leq 0 \\
\frac{\sin(3x)}{2x + 1} = 0
\end{align*}
$$
\
\
\
$$
f(x) = \frac{x^2 - 1}{x^2 + 1}
$$
I will be diving both sides by $ \frac{1}{x^2}$
\
$$
f(x) = \frac{1 - \frac{1}{x^2}}{1 + \frac{1}{x^2}} = \frac{1-0}{1+0} = 1
$$
\
\
\
$$
\begin{align*}
g(x) = \frac{\sin(x)}{x^2+1} \\
-1 \leq \frac{sin(x)}{x^2 + 1} \leq 1  \\
\frac{-1}{x^2 + 1} \leq \frac{sin(x)}{x^2 + 1} \leq \frac{1}{x^2 + 1}
\end{align*}
$$
Either of the sides will eventually equal to 0
\
$$
\frac{-1}{x^2 + 1} \leq \frac{sin(x)}{x^2 + 1} \leq \frac{1}{x^2 + 1}
$$
\
$$
0 \leq \frac{sin(x)}{x^2 + 1} \leq 0
$$
The answer is 0
